```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the trajectories with one peak
xtrainx1 = np.loadtxt('Xtrainx1_1.csv', delimiter=',')
```

```python
#configure font properties for the plot
plt.rc('font',family='Times New Roman')
plt.rcParams.update({'font.size':13})
plt.rcParams['font.size'] = 13
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'
plt.rcParams['axes.linewidth'] = 1
```

```python
#plot 25 of the trajectories with one peak
t = np.linspace(0, 50, 800)
plt.plot(t, xtrainx1[0:25,:].T, linewidth = 0.3)
plt.xlabel(r"$\mathrm{\omega_0 t}$")
plt.ylabel(r"$\mathrm{\langle \sigma_x (t) \rangle}$")
plt.savefig('trainingsetoneRC.pdf')
plt.show()
```
